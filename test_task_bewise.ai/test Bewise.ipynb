```python
df=df[df['role']=='manager']
df = df.reset_index(drop=True)
df.text = df.text.str.lower()
```

```python
results = pd.DataFrame(columns=['id разговора', 'Поздоровался', 'Представился(реплика)', 
                                'Назвал имя', 'Назвал компанию', 'Попрощался', 'И попрощался и поздоровался'], 
                       index=range(6))
results
```

```python
for id in set(df.dlg_id):
    a = df[df['dlg_id']<id+1]
    a = a[a['dlg_id']>=id] 
    b = []
    sliced_df = a.head(2)
    sliced_df = sliced_df.append(df.tail(2))
    sliced_df
    greeting = 0
    introduce = 0
    mng_name = 0
    company = 0
    farewell = 0
    results['id разговора'][id]=id
    for i in sliced_df.text:
        b.append(i.split())
    for i in b:
        if 'здравствуйте' in i:
            print('Менеджер, с id разговора {} поздоровался фразой:'.format(id), *i)
            greeting+=1
            print()
            results['Поздоровался'][id]='Да'
        elif 'добрый' and 'день' in i:
            print('Менеджер, с id разговора {} поздоровался фразой:'.format(id), *i)
            greeting+=1
            print()
            results['Поздоровался'][id]='Да'
        elif 'меня' in i and 'зовут' in i:
            print('Менеджер, с id разговора {} представился фразой:'.format(id), *i)
            print()
            introduce+=1
            results['Представился(реплика)'][id]='Да'
            for name in all_names:
                if name in i:
                    print('Менеджерa, с id разговора {} зовут:'.format(id), name)
                    print()
                    mng_name+=1
                    results['Назвал имя'][id]='Да'
            for k in b:
                for kk in range(len(k)):
                    if k[kk]=='компания':
                        print('Менеджерa, с id разговора {} работает в компании:'.format(id), k[kk+1])
                        print()
                        company+=1
                        results['Назвал компанию'][id]='Да'
        elif 'хорошего' in i:
            print('Менеджер, с id разговора {} попрощался фразой:'.format(id), *i)
            farewell+=1
            results['Попрощался'][id]='Да'
            print()
        elif 'до' in i and 'свидания' in i:
            print('Менеджер, с id разговора {} попрощался фразой:'.format(id), *i)
            farewell+=1
            results['Попрощался'][id]='Да'
            print()
    if greeting>=1 and farewell>=1:
        print('Менеджер, с id разговора {} поздоровался и попрощался с клиентом'.format(id))
        results['И попрощался и поздоровался'][id]='Да'
    if greeting<1:
        print('Менеджер, с id разговора {} не поздоровался с клиентом'.format(id))
    if introduce<1:
        print('Менеджер, с id разговора {} не представился'.format(id))
    if mng_name<1:
        print('Менеджер, с id разговора {} не назвал свое имя'.format(id))
    if company<1:
        print('Менеджер, с id разговора {} не назвал свою компанию'.format(id))
    if farewell<1:
        print('Менеджер, с id разговора {} не назвал не попрощался с клиентом'.format(id))
    print()
    print()
    print()
    
```
